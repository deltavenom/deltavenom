```python
data = df.iloc[1: , :]
data = data.drop(columns=["Storages","Display"])
data.head()
```

```python
def get_name(value):
    v1 = value.split(' (')
    name = v1[0]
    return name


def get_color(value):
    v1 = value.split('(')
    if len(v1)>=2:
        v2 = v1[1].split(',')
        color = v2[0]
    else:
        color = np.nan
    return color
```

```python
data["Model Name"] = data["Model_Name"].apply(get_name)
data["Color"] = data["Model_Name"].apply(get_color)
```

```python
def replac(value):
    try:
        value = value.replace("₹","")
        value = value.replace(",","")
        return int(value)
    except:
        return value
```

```python
data["price"] = data["Price"].apply(replac)
```
